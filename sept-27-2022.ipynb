```python
# Tidy tuesday 9/27
# https://github.com/rfordatascience/tidytuesday/tree/master/data/2022/2022-09-27
# Location quotients (LQ) measure an artist occupation's concentration in the labor force, relative to the U.S. 
# labor force share. For example, an LQ of 1.2 indicates that the state's labor force in an occupation is 20 percent
# greater than the occupation's national labor force share. An LQ of 0.8 indicates that the state's labor force 
#in an occupation is 20 percent below the occupation's national labor force share.

import pandas as pd 
import numpy as np 
import plotly.express as px
```

```python
artists = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-09-27/artists.csv")
```

```python
counts_by_state = artists[["state", "artists_n"]]

counts_by_state = counts_by_state.dropna()

counts_by_state_grouped = counts_by_state.groupby(["state"]).sum()

counts_by_state_sorted = counts_by_state_grouped.sort_values(by="artists_n", ascending=False)
```
